```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['is_declined_n']=le.fit_transform(df['Is_Declined'])
df['Is_Foreign_Transaction_n']=le.fit_transform(df['Is_Foreign_Transaction'])
df['Is_HighRisk_Country_n']=le.fit_transform(df['Is_HighRisk_Country'])
df['Is_Fradulent_n']=le.fit_transform(df['Is_Fradulent'])
df
```

```python
df=df.drop(['Merchant_Id','Is_Declined','Is_Foreign_Transaction','Is_HighRisk_Country','Is_Fradulent'],axis='columns')
df
```

```python
df=df.rename(columns={'6_Month_Avg_Chbk_Amt': 'Month_Avg_Chbk_Amt', '6_Month_Chbk_Freq': 'Month_Chbk_Freq'})
df
```

```python
x=df.iloc[:,:9]
x
```

```python
y=df.iloc[:,9]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
rfc.predict([[100.0,3000.0,5,0.0,0.0,0,0,1,1]])
```
